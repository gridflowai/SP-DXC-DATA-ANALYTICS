---------------------------
#### Curse of dimensionality
-----------------------------

```python
import numpy as np

from sklearn.metrics.pairwise import euclidean_distances
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data with varying dimensions
num_points     = 50
max_dimensions = 100

data = np.random.rand(num_points, max_dimensions)
```

```python
# Calculate Euclidean distances for different numbers of dimensions
num_dimensions = list(range(1, max_dimensions + 1))
distances      = []

for dim in num_dimensions:
    dist_matrix   = euclidean_distances(data[:, :dim])
    mean_distance = np.mean(dist_matrix)
    distances.append(mean_distance)
```

```python
# Plotting
plt.plot(num_dimensions, distances, marker='o')
plt.xlabel('Number of Dimensions')
plt.ylabel('Mean Euclidean Distance')
plt.title('Curse of Dimensionality')
plt.show()
```

**Summary**

- we generate synthetic data with varying dimensions and then calculate the mean Euclidean distance for each subset of dimensions. 
- As the number of dimensions increases, the mean Euclidean distance tends to grow, illustrating the curse of dimensionality. 
- You will likely observe that the distances increase rapidly as the number of dimensions becomes larger, showing that the notion of distance becomes less meaningful in high-dimensional spaces.

#### Issue of equidistance

```python
# Generate random points in a 2D space
num_points = 100
points = np.random.rand(num_points, 2)
```

```python
# Calculate pairwise distances
distances = np.sqrt(np.sum((points[:, np.newaxis] - points) ** 2, axis=2))
```

```python
# Plotting
plt.figure(figsize=(10, 4))

plt.subplot(121)
plt.scatter(points[:, 0], points[:, 1], c='blue')
plt.title('Original 2D Space')

plt.subplot(122)
plt.imshow(distances, cmap='viridis', interpolation='none')
plt.colorbar(label='Distance')
plt.title('Pairwise Distances')
plt.xticks([])
plt.yticks([])

plt.tight_layout()
```

**Summary**
- we generate random points in a 2D space and calculate the pairwise distances between them. 
- The left subplot shows the original 2D space with points scattered randomly. 
- The right subplot displays a heatmap of the pairwise distances, where darker regions indicate shorter distances.

- Equidistance becomes more pronounced as the number of dimensions increases, which is difficult to visualize directly. 
- In high-dimensional spaces, the distances between points tend to become more uniform, leading to the phenomenon of equidistance.
