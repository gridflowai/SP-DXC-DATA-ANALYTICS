### Euclidean Distance Formula
#### Definition:
Euclidean distance is a measure of the straight-line distance between two points in a two-dimensional or multi-dimensional space. It is calculated using the Pythagorean theorem and is widely used in various fields, including geometry, machine learning, and data analysis.

#### Explanation:
The Euclidean distance formula for two points A and B in a two-dimensional space is given by:

distance = √((x₂ - x₁)² + (y₂ - y₁)²)


For three-dimensional space, the formula extends to:

distance = √((x₂ - x₁)² + (y₂ - y₁)² + (z₂ - z₁)²)


#### Examples:
For two points A(2, 3) and B(5, 7) in the 2D plane:

distance = √((5 - 2)² + (7 - 3)²)
         = √(3² + 4²)
         = √(9 + 16)
         = √25
         = 5


For two points C(1, 2, 3) and D(4, 5, 6) in 3D space:

distance = √((4 - 1)² + (5 - 2)² + (6 - 3)²)
         = √(3² + 3² + 3²)
         = √(9 + 9 + 9)
         = √27
         ≈ 5.196


#### Python impl

```python
import numpy as np
```

```python
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point2 - point1) ** 2))
```

```python
# Example 1 - 2D space
point_a = np.array([2, 3])
point_b = np.array([5, 7])
distance_2d = euclidean_distance(point_a, point_b)
print("Euclidean distance in 2D:", distance_2d)

# Example 2 - 3D space
point_c = np.array([1, 2, 3])
point_d = np.array([4, 5, 6])
distance_3d = euclidean_distance(point_c, point_d)
print("Euclidean distance in 3D:", distance_3d)
```

#### with a plot

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
def euclidean_distance(point1, point2):
    return np.linalg.norm(point2 - point1)
```

```python
# Example 1 - 2D space
point_a_2d = np.array([2, 3])
point_b_2d = np.array([5, 7])

distance_2d = euclidean_distance(point_a_2d, point_b_2d)
```

```python
# Plotting 2D points and distance
plt.figure(figsize=(6, 4))
plt.plot([point_a_2d[0], point_b_2d[0]], [point_a_2d[1], point_b_2d[1]], marker='o')
plt.text(point_a_2d[0], point_a_2d[1], 'A', fontsize=12, ha='right')
plt.text(point_b_2d[0], point_b_2d[1], 'B', fontsize=12, ha='right')
plt.annotate(f"Distance: {distance_2d:.2f}", xy=(3.5, 5), fontsize=12)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Euclidean Distance in 2D')
plt.grid(True)
plt.show()
```

```python
# Example 2 - 3D space
point_c_3d = np.array([1, 2, 3])
point_d_3d = np.array([4, 5, 6])
distance_3d = euclidean_distance(point_c_3d, point_d_3d)
```

```python
# Plotting 3D points and distance
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(point_c_3d[0], point_c_3d[1], point_c_3d[2], marker='o', label='C')
ax.scatter(point_d_3d[0], point_d_3d[1], point_d_3d[2], marker='o', label='D')
ax.plot([point_c_3d[0], point_d_3d[0]], [point_c_3d[1], point_d_3d[1]], [point_c_3d[2], point_d_3d[2]], label='Distance')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Euclidean Distance in 3D')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate synthetic data with make_blobs
n_samples = 20
n_features = 2
n_classes = 3
X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_classes, random_state=42)

# Class labels
class_labels = ['Class A', 'Class B', 'Class C']

# Create a 2D scatter plot
plt.figure(figsize=(8, 6))

# Plot each class with different colors
for i in range(n_classes):
    class_indices = np.where(y == i)[0]
    class_data = X[class_indices]
    plt.scatter(class_data[:, 0], class_data[:, 1], label=class_labels[i], marker='o')

# Set labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Bird Classification using make_blobs')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

####  Raw Euclidean Distance

Computes the Euclidean distance between two 1-D arrays.

The Euclidean distance between 1-D arrays u and v, is defined as

${\| u - v\|}_2$ = $\sum (w_i\mid(u_i - v_i)\mid^2) $

** key points **

- In 2D or 3D the euclidean distance will be a straight line

- properties
    - d(i, j) >= 0
    - d(i, j) = 0
    - d(i, j) = d(j, i)
    - d(i, j) <= d(i, k) + d(k, j)

- Pythagoras’ theorem extends to vectors in multidimensional space 

- squared distances between two vectors in multidimensional space are the sum of squared differences in their coordinates. This multidimensional distance is called the Euclidean distance

3. When variables are on different measurement scales, __standardization__ is necessary to balance the contributions of the variables in the computation of distance. The Euclidean distance computed on standardized variables is called the __standardized Euclidean distance.__

4. __weighted Euclidean distance.__

5. A particular weighted Euclidean distance applicable to count data is the chi-square distance, which is calculated between the relative counts for each sample, called profiles, and weights each variable by the inverse of the variable’s overall mean count.

#### scipy - euclidean distance

- `euclidean_distances` function from the `scipy.spatial.distance` module is used to calculate the pairwise Euclidean distances between sets of points. 

```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
```

```python
# Example 1: Calculating Euclidean distances between 2D points
points = np.array([[1, 2],
                   [4, 6],
                   [2, 3]])
```

```python
euclidean_distances(points)
```

```python
# Example 2: Calculating Euclidean distances between 3D points
points_3d = np.array([[1, 2, 3],
                      [4, 6, 5],
                      [2, 3, 1]])

euclidean_distances(points_3d)
```
