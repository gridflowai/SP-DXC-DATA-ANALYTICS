#### points, line, equation

```python
import matplotlib.pyplot as plt
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __str__(self):
        return f"({self.x}, {self.y})"
    
class Line:
    def __init__(self, slope, intercept):
        self.slope = slope
        self.intercept = intercept
        
    def equation(self):
        return f"y = {self.slope}x + {self.intercept}"
        
    def __str__(self):
        return self.equation()
```

```python
# Creating points
point1 = Point(2, 3)
point2 = Point(5, 8)

print("Point 1:", point1)
print("Point 2:", point2)
```

```python
# Creating a line
line = Line(2, 1)

print("Line:", line)
```

```python
# Calculate distance between two points
distance = ((point2.x - point1.x)**2 + (point2.y - point1.y)**2)**0.5
print("Distance between points:", distance)
```

```python
# Calculate slope and intercept of a line passing through two points
def calculate_line(point1, point2):
    slope = (point2.y - point1.y) / (point2.x - point1.x)
    intercept = point1.y - slope * point1.x
    return Line(slope, intercept)
```

```python
line_through_points = calculate_line(point1, point2)
print("Line passing through points:", line_through_points)
```

```python
# Plotting
plt.figure(figsize=(6, 4))

# Plotting points
plt.scatter([point1.x, point2.x], [point1.y, point2.y], color='red', label='Points')
plt.text(point1.x, point1.y, 'Point 1', fontsize=12, verticalalignment='bottom')
plt.text(point2.x, point2.y, 'Point 2', fontsize=12, verticalalignment='bottom')

# Plotting line
x_vals = [point1.x, point2.x]
y_vals = [line_through_points.slope * x + line_through_points.intercept for x in x_vals]
plt.plot(x_vals, y_vals, color='blue', label=line_through_points.equation())

# Adding distance as text on the plot
plt.text((point1.x + point2.x) / 2, (point1.y + point2.y) / 2, f'Distance: {distance:.2f}', fontsize=12, verticalalignment='top')


plt.xlabel('X')
plt.ylabel('Y')
plt.xlim((0,20))
plt.ylim((0,20))
plt.title('Points, Line, and Equation')
plt.legend()
plt.grid(True)
plt.show()
```

#### Curves etc

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate points for the curves
x = np.linspace(-5, 5, 400)

# Circle equation: (x - h)^2 + (y - k)^2 = r^2
h, k, r = 1, 2, 3
y_circle_positive = np.sqrt(r**2 - (x - h)**2) + k
y_circle_negative = -np.sqrt(r**2 - (x - h)**2) + k

# Ellipse equation: (x-h)^2/a^2 + (y-k)^2/b^2 = 1
a, b = 3, 2
y_ellipse_positive = b * np.sqrt(1 - (x - h)**2 / a**2) + k
y_ellipse_negative = -b * np.sqrt(1 - (x - h)**2 / a**2) + k

# Parabola equation: y = ax^2 + bx + c
a_parabola, b_parabola, c_parabola = 0.5, -1, 2
y_parabola = a_parabola * x**2 + b_parabola * x + c_parabola

# Hyperbola equation: (x-h)^2/a^2 - (y-k)^2/b^2 = 1
h_hyperbola, k_hyperbola, a_hyperbola, b_hyperbola = 0, 0, 1, 2
y_hyperbola_positive = (b_hyperbola / a_hyperbola) * np.sqrt((x - h_hyperbola)**2 / a_hyperbola**2 - 1) + k_hyperbola
y_hyperbola_negative = -(b_hyperbola / a_hyperbola) * np.sqrt((x - h_hyperbola)**2 / a_hyperbola**2 - 1) + k_hyperbola
```

```python
# Plotting
plt.figure(figsize=(7, 5))

# Circle plot
plt.plot(x, y_circle_positive, label='Circle')
plt.plot(x, y_circle_negative)
plt.scatter([h], [k], color='red')  # Center point

# Ellipse plot
plt.plot(x, y_ellipse_positive, label='Ellipse')
plt.plot(x, y_ellipse_negative)
plt.scatter([h], [k], color='red')  # Center point

# Parabola plot
plt.plot(x, y_parabola, label='Parabola')

# Hyperbola plot
plt.plot(x, y_hyperbola_positive, label='Hyperbola')
plt.plot(x, y_hyperbola_negative)
plt.scatter([h_hyperbola], [k_hyperbola], color='red')  # Center point

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Equations of Curves')
plt.legend()
plt.grid(True)
plt.show()
```

#### Circle:

- `Data Clustering`: 

    - In unsupervised learning, you might use clustering algorithms like K-means. If the data clusters resemble circular shapes, you can visualize the clusters using circles centered around cluster centers.

- `Anomaly Detection`: 

    - Circles can also be used in anomaly detection to define a region of normal behavior. Data points outside the circle might be considered outliers.

**Data clustering**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate synthetic data with circular clusters
np.random.seed(0)

# First cluster (centered at [2, 2])
cluster1 = np.random.randn(50, 2) + [2, 2]

# Second cluster (centered at [8, 8])
cluster2 = np.random.randn(50, 2) + [8, 8]

# Combine the clusters
data = np.vstack([cluster1, cluster2])

# Apply K-means clustering
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(data)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Plot the clusters and cluster centers
plt.figure(figsize=(6, 4))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering with Circular Clusters')
plt.legend()
plt.grid(True)
plt.show()
```

**Anomaly detection**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data points
np.random.seed(0)
normal_data = np.random.normal(0, 1, (100, 2))
anomaly_data = np.random.normal(5, 1, (10, 2))

# Center and radius of the circle for normal behavior
circle_center = (0, 0)
circle_radius = 2

# Plotting
plt.figure(figsize=(6, 4))

# Plot normal data points inside the circle
plt.scatter(normal_data[:, 0], normal_data[:, 1], color='blue', label='Normal Data')

# Plot anomaly data points outside the circle
plt.scatter(anomaly_data[:, 0], anomaly_data[:, 1], color='red', label='Anomaly Data')

# Plot the detection circle
circle = plt.Circle(circle_center, circle_radius, color='green', fill=False, label='Anomaly Detection Circle')
plt.gca().add_patch(circle)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Anomaly Detection using Circles')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

- In this code, we generate random data points representing normal and anomaly behavior. 
- The normal data points are generated from a standard normal distribution (centered at 0 with a standard deviation of 1), while the anomaly data points are generated from a shifted normal distribution (centered at 5 with a standard deviation of 1).

- We define a circle with a specified center and radius to represent the region of normal behavior. 
- `Data points outside this circle are considered anomalies`. 
- The code then plots the normal and anomaly data points along with the anomaly detection circle.

#### Ellipse:

- `Covariance Ellipses`: 
    - In multivariate statistics, covariance ellipses depict the spread and correlation between dimensions. Principal Component Analysis (PCA) often uses ellipses to show the directions of maximum variance.
    
- `Confidence Regions`: 
    - In regression analysis, ellipses can represent confidence regions for predicted outcomes, indicating the uncertainty in predictions.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Generate random data with covariance
np.random.seed(0)
mean = [0, 0]

cov  = [[2, 1  ], 
        [1, 1.5]]

data = np.random.multivariate_normal(mean, cov, 100)

# Perform PCA
pca = PCA(n_components=2)
pca.fit(data)

pca_components           = pca.components_
explained_variance_ratio = pca.explained_variance_ratio_

# Create covariance ellipses
def plot_cov_ellipse(mean, cov, ax, color='black', alpha=0.3):
    v, w = np.linalg.eigh(cov)
    v = 2.0 * np.sqrt(2.0) * np.sqrt(v)
    u = w[0] / np.linalg.norm(w[0])
    angle = np.arctan(u[1] / u[0])
    angle = 180.0 * angle / np.pi
    ell   = plt.matplotlib.patches.Ellipse(mean, v[0], v[1], 180.0 + angle, linewidth=5, color=color, alpha=alpha)
    ax.add_artist(ell)
    return ell

# Plotting
plt.figure(figsize=(9, 4))
plt.scatter(data[:, 0], data[:, 1], color='blue', s=1, label='Data')

# Plot covariance ellipses
for i, (explained_var, component) in enumerate(zip(explained_variance_ratio, pca_components)):
    ellipse = plot_cov_ellipse(pca.mean_, np.outer(component, component) * explained_var, plt.gca(), color='red')
    ellipse.set_label(f'Covariance Ellipse {i+1} ({explained_var:.2f} explained variance)')

plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Covariance Ellipses using PCA')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

we generate random data with a specified mean and covariance matrix. We then use PCA to compute the principal components and explained variance ratio. For each principal component, we create covariance ellipses based on the component and explained variance.

The function plot_cov_ellipse is used to plot the covariance ellipses. It calculates the ellipse parameters based on the eigenvectors and eigenvalues of the covariance matrix.

`Confidence Regions:`

In regression analysis, ellipses can represent confidence regions for predicted outcomes, indicating the uncertainty in predictions.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(0)
X = np.random.rand(50, 1) * 10
y = 2 * X + 1 + np.random.randn(50, 1) * 2

# Fit linear regression model
reg = LinearRegression()
reg.fit(X, y)

# Compute prediction and residuals
X_pred = np.linspace(0, 10, 100).reshape(-1, 1)
y_pred = reg.predict(X_pred)
residuals = y - reg.predict(X)

# Compute confidence regions
confidence_level = 0.95
alpha = 1 - confidence_level
mse = np.mean(residuals**2)
std_error = np.sqrt(mse)

# Calculate confidence intervals
confidence_intervals = std_error * np.sqrt(1 + 1 / len(X_pred) + (X_pred - np.mean(X))**2 / np.sum((X - np.mean(X))**2))

# Plotting
plt.figure(figsize=(6, 4))
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X_pred, y_pred, color='red', label='Regression Line')

# Confidence regions
plt.fill_between(X_pred[:, 0], (y_pred - confidence_intervals).flatten(), (y_pred + confidence_intervals).flatten(), color='gray', alpha=0.3, label=f'{confidence_level*100}% Confidence Region')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Confidence Regions in Regression Analysis')
plt.legend()
plt.grid(True)
plt.show()
```
