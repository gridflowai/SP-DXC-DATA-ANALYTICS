```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Coefficients A, B, and C for the standard form Ax + By = C
A = 2
B = -3
C = 6
```

```python
# Generate x values for plotting
x_values = np.linspace(-10, 10, 100)
```

```python
# Calculate corresponding y values using the standard form equation
# y = (C - Ax) / B
y_values = (C - A * x_values) / B
```

```python
# Plotting the line
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'{A}x + {B}y = {C}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line in Standard Form')
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.grid(True)
plt.legend()
plt.xlim(-10, 10)
plt.ylim(-10, 10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate synthetic data with a positive linear relationship
np.random.seed(0)
age = np.linspace(20, 70, 20)
income = 2000 + 50 * age + np.random.randn(20) * 100
```

```python
# Reshape the features
age = age.reshape(-1, 1)
```

```python
# Fit a linear regression model
model = LinearRegression()
model.fit(age, income)
```

```python
# Plot the linear regression line
age_range = np.linspace(20, 70, 100)
income_pred = model.predict(age_range.reshape(-1, 1))
plt.scatter(age, income, label='Data Points')
plt.plot(age_range, income_pred, color='red', label='Linear Regression Line')

plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Positive Linear Relationship between Age and Income')
plt.legend()
plt.grid(True)
```
