```python
import matplotlib.pyplot as plt

# Create figure and axis
fig, ax = plt.subplots()

# Set limits for the coordinate plane
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

# Add grid lines
ax.grid(True, which='both')

# Set x and y axis labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add x and y axis arrows
ax.annotate('', xy=(9.5, 0), xytext=(-9.5, 0),
            arrowprops=dict(arrowstyle='->', lw=1.5))
ax.annotate('', xy=(0, 9.5), xytext=(0, -9.5),
            arrowprops=dict(arrowstyle='->', lw=1.5))

# Mark the origin (0, 0)
ax.plot(0, 0, 'ro', markersize=5, label='Origin (0, 0)')

# Add vertical and horizontal lines at x = 0 and y = 0
ax.axvline(x=0, color='black', linewidth=0.8, linestyle='--', label='x = 0')
ax.axhline(y=0, color='black', linewidth=0.8, linestyle='--', label='y = 0')

# Set aspect ratio to be equal
ax.set_aspect('equal', adjustable='datalim')

# Add legend
ax.legend()

# Set plot title
plt.title('Coordinate Plane with Vertical and Horizontal Lines')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create figure and axis
fig, ax = plt.subplots()

# Set limits for the coordinate plane
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

# Add grid lines
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

# Set x and y axis labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add x and y axis arrows
ax.annotate('', xy=(9.5, 0), xytext=(-9.5, 0),
            arrowprops=dict(arrowstyle='->', lw=1.5))
ax.annotate('', xy=(0, 9.5), xytext=(0, -9.5),
            arrowprops=dict(arrowstyle='->', lw=1.5))

# Mark the origin (0, 0)
ax.plot(0, 0, 'ro', markersize=5, label='Origin (0, 0)')

# Add vertical and horizontal lines at x = 0 and y = 0
ax.axvline(x=0, color='black', linewidth=0.8, linestyle='--', label='x = 0')
ax.axhline(y=0, color='black', linewidth=0.8, linestyle='--', label='y = 0')

# Add labels for the quadrants
ax.text(2, 2, 'Quadrant I', fontsize=10, color='blue')
ax.text(-8, 2, 'Quadrant II', fontsize=10, color='blue')
ax.text(-8, -4, 'Quadrant III', fontsize=10, color='blue')
ax.text(2, -4, 'Quadrant IV', fontsize=10, color='blue')

# Points
points = [(2, 4), (-2, -4)]
for point in points:
    ax.plot(point[0], point[1], 'go', markersize=5)
    ax.annotate(f'({point[0]}, {point[1]})', xy=point, textcoords='offset points', xytext=(10, 10), fontsize=8, color='green')

    # Plot perpendicular lines
    ax.plot([point[0], point[0]], [0, point[1]], 'g--', linewidth=0.8)
    ax.plot([0, point[0]], [point[1], point[1]], 'g--', linewidth=0.8)
    
    # Labels for "abscissa" and "ordinate"
    ax.text(point[0], 0, 'abscissa', fontsize=8, color='black', ha='center', va='bottom')
    ax.text(0, point[1], 'ordinate', fontsize=8, color='black', ha='right', va='center')

# Set aspect ratio to be equal
ax.set_aspect('equal', adjustable='datalim')

# Add legend
ax.legend()

# Set plot title
plt.title('Coordinate Plane with Axes, Origin, Quadrants, Points, and Lines')

# Show the plot
plt.show()
```
