#### Euclidean
Our euclidean distance function can be defined as follows:

$$ \Large \sqrt{\sum^n_{i=1} (x_i - y_i)^2} $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Coordinates of the points
A = (1, 2)
B = (4, 6)

# Plot the points
plt.scatter([A[0], B[0]], [A[1], B[1]], c=['red', 'blue'], marker='o')
plt.text(A[0], A[1], 'A(1, 2)', ha='right')
plt.text(B[0], B[1], 'B(4, 6)', ha='right')

# Plot the line connecting the points (representing Euclidean distance)
plt.plot([A[0], B[0]], [A[1], B[1]], linestyle='--', color='green')

# Highlight the distance value
plt.text((A[0] + B[0]) / 2, (A[1] + B[1]) / 2, f'Euclidean Distance = {5:.2f}', ha='center', color='green')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Euclidean Distance Example')
plt.grid(True)
plt.show()
```

$$ d_{2}(x)=\sqrt{(x_2 - x_1)^{2} +\left(y_{2}-y_{1}\right)^{2}}$$

```python
u = np.array([1,2])
v = np.array([4,5])
```

```python
euc_dist = np.sqrt((u[0]- v[0])**2 + (u[1]- v[1])**2 )
euc_dist
```

$$ d_{3}(x)=\sqrt{(z_2 - z_1)^{2} + (x_2 - x_1)^{2} +\left(y_{2}-y_{1}\right)^{2}}$$

```python
u = np.array([1,2,8])
v = np.array([4,5,9])
```

```python
euc_dist = np.sqrt((u[0]- v[0])**2 + (u[1]- v[1])**2 + (u[2]- v[2])**2)
euc_dist
```

generalized function

```python
def euclidean_distance(u, v):   
    return np.sqrt(np.sum((u - v) ** 2))
```

```python
u = np.array([1,2,8])
v = np.array([4,5,9])
```

```python
euclidean_distance(u, v)
```

#### using scipy

```python
from scipy.spatial import distance

u = np.array([1,2,8])
v = np.array([4,5,9])

d = distance.euclidean(u, v)
print("Euclidean distance: ",d)
```

#### using numpy

```python
dist = np.linalg.norm(u - v, ord=2)
print("Euclidean distance: ", dist)
```

```python
import pandas as pd
```

```python
data = {'name':            ['rahim', 'bhupen', 'raja', 'gaba', 'kiran', 'kia', 'jalebi'],
        'number_of_loans': [4,        1,        3,       5,       4,     4,      6     ], 
        'salary':          [80000,   60000,    45000,  120000,  30000,  45000,  55000  ], 
        'expenses':        [70000,   30000,    40000,  67000,    25000, 15000,  54000]}
```

```python
df = pd.DataFrame(data)
df
```

scale the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['number_of_loans', 'salary', 'expenses']])
```

```python
scaled_df = scaler.transform(df[['number_of_loans', 'salary', 'expenses']])
```

```python
scaled_df
```

scatter plot in 3D

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax  = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(scaled_df[:, 0], scaled_df[:, 1], scaled_df[:, 2], color = "black", s=100, alpha=.5)
ax.set_xlabel('# loans', fontweight ='bold')
ax.set_ylabel('salary', fontweight ='bold')
ax.set_zlabel('Expenses', fontweight ='bold')
ax.set_title("Customer profile");

scaled_df[:, 0], scaled_df[:, 1], scaled_df[:, 2], df['name'].values

def label_point(x, y, z, val, ax):
    a = pd.concat({'x': x, 'y': y, 'z': z, 'val': val}, axis=1)
    
    for i, point in a.iterrows():
        ax.text(point['x'], 
                point['y'], 
                point['z'], 
                point['val'], 
                c='r', 
            
                family='sans-serif', 
                fontsize=12, )

label_point(pd.Series(scaled_df[:, 0]), pd.Series(scaled_df[:, 1]), pd.Series(scaled_df[:, 2]), df['name'], ax)
```
