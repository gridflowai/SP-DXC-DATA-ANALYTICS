```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
import matplotlib.pyplot as plt
```

```python
# Generate random data with an outlier
np.random.seed(42)
num_points  = 10
data        = np.random.rand(num_points, 2)

outlier     = np.array([10, 10])

data_with_outlier = np.vstack((data, outlier))
```

```python
# Calculate Euclidean distances
distances_with_outlier = euclidean_distances(data_with_outlier)
```

```python
# Plotting
plt.figure(figsize=(10, 4))

plt.subplot(121)
plt.scatter(data[:, 0], data[:, 1], label='Data')
plt.scatter(outlier[0], outlier[1], color='red', label='Outlier')
plt.legend()
plt.title('Data with Outlier')

plt.subplot(122)
plt.imshow(distances_with_outlier, cmap='viridis', interpolation='none')
plt.colorbar(label='Distance')
plt.title('Pairwise Distances with Outlier')
plt.xticks([])
plt.yticks([])

plt.tight_layout()
```

- we generate random 2D data points and add an outlier point far away from the rest of the data. 
- We then calculate the pairwise Euclidean distances between all points (including the outlier) and create a heatmap to visualize the distances.

- observe that the outlier significantly increases the distances between itself and all other points. 

- This illustrates how outliers can distort distance calculations and impact the overall interpretation of relationships between data points. 

- Outliers can have a strong influence on algorithms that rely on distance metrics, potentially leading to biased results or inaccurate clustering/classification.
