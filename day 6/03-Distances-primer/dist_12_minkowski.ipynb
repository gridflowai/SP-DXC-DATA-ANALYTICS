#### Minkowski distance with a custom value of "p":
------------------------------

```python
import numpy as np
from scipy.spatial.distance import minkowski
```

```python
# Define two points in a 3-dimensional space
point1 = np.array([1, 2, 3])
point2 = np.array([4, 6, 5])
```

```python
# Calculate the Minkowski distance with custom p value
p = 3  # Change p to try different Minkowski distances
minkowski_distance = minkowski(point1, point2, p)

print(f"Minkowski Distance (p={p}): {minkowski_distance}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the points
point1 = np.array([1, 2])
point2 = np.array([3, 4])

# Define p values
p_values = [1, 2, 3, 4, 5]

# Create a plot
plt.figure(figsize=(12, 6))

# Plot the points
plt.plot(point1[0], point1[1], 'go', label='Point 1 (1, 2)')
plt.plot(point2[0], point2[1], 'bo', label='Point 2 (4, 5)')

colors = ['b', 'c', 'g', 'r', 'k']
# Plot Minkowski distances for different p values
for p in p_values:
    distance = np.linalg.norm(point2 - point1, ord=p)
    
    plt.axhline(y=distance, c = colors[p-1], label="p = {}".format(p) )
    #plt.annotate(f'p={p}: {distance:.2f}', (2, 8-p), color='red', fontsize=10)
    plt.arrow(point1[0], point1[1], point2[0] - point1[0], point2[1] - point1[1], 
              head_width=0.15, head_length=0.15, fc='gray', ec='gray', linestyle='--')
    
    # Highlight distances for p=1 and p=2
    # if p in [1, 2]:
    #     plt.text(point1[0] + (point2[0] - point1[0]) / 2, 
    #              point1[1] + (point2[1] - point1[1]) / 2 + 0.3, f'{distance:.2f}', color='blue')

# Set labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Geometric Intuition of Minkowski Distance')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
def custom_minkowski_distance(point1, point2, p):
    if len(point1) != len(point2):
        raise ValueError("Points must have the same number of dimensions")
    
    distance_sum = sum(abs(coord2 - coord1)**p for coord1, coord2 in zip(point1, point2))
    distance = distance_sum ** (1/p)
    return distance

# Define the points
point1 = (1, 2)
point2 = (3, 4)

# Define the value of p
p = 2

# Calculate the custom Minkowski distance
distance = custom_minkowski_distance(point1, point2, p)
print(f"Custom Minkowski distance between {point1} and {point2} with p = {p}: {distance:.4f}")
```
