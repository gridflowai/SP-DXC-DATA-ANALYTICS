#### Optimization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the constraints
x = np.linspace(0, 10, 400)
y1 = 8 - 2*x
y2 = (6 + x) / 2

# Create the plot
plt.figure(figsize=(8, 4))

# Plot the constraints
plt.plot(x, y1, label=r'$2x + y \leq 8$')
plt.plot(x, y2, label=r'$-x + 2y \leq 6$')
plt.fill_between(x, np.minimum(y1, y2), color='gray', alpha=0.3, label='Feasible Region')

# Plot the objective function
plt.plot(x, (12 - 3*x) / 4, label=r'$3x + 4y$')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Optimization: Feasible Region and Minimize')
plt.xlim((0, 8))
plt.ylim((0, 5))

# Add legend
plt.legend()

# Show the plot
plt.show()
```

#### Gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt

# Objective function: f(x) = x^2
def objective_function(x):
    return x**2

# Gradient of the objective function: f'(x) = 2x
def gradient(x):
    return 2 * x

# Gradient Descent function
def gradient_descent(initial_x, learning_rate, num_iterations):
    x_values = [initial_x]
    for _ in range(num_iterations):
        x_new = x_values[-1] - learning_rate * gradient(x_values[-1])
        x_values.append(x_new)
    return x_values

# Hyperparameters
initial_x = 4  # Initial guess for x
learning_rate = 0.1
num_iterations = 10

# Perform gradient descent
x_values = gradient_descent(initial_x, learning_rate, num_iterations)

# Plot the objective function and the gradient descent path
x = np.linspace(-5, 5, 400)
y = objective_function(x)
plt.plot(x, y, label='Objective Function: f(x) = x^2')

# Plot gradient descent path
plt.scatter(x_values, objective_function(np.array(x_values)), color='red', label='Gradient Descent Path')

# Plot lines representing gradients at each step
for x_val in x_values[:-1]:
    tangent_line = gradient(x_val) * (x - x_val) + objective_function(x_val)
    plt.plot(x, tangent_line, linestyle='--', color='green', alpha=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Example with Tangent Lines')
plt.legend()
plt.grid()
plt.show()
```

#### Outlier detection

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data with a linear trend
np.random.seed(0)
X = np.random.rand(50, 1) * 10
y = 2 * X + 3 + np.random.randn(50, 1)

# Introduce an outlier
outlier_index = 25
y[outlier_index] += 20

# Fit a linear regression line to the data
regression = LinearRegression()
regression.fit(X, y)

# Predict the expected values using the regression line
y_pred = regression.predict(X)

# Calculate the residuals (deviations from the regression line)
residuals = y - y_pred

# Plot the data points, regression line, and residuals
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Data points', color='blue')
plt.plot(X, y_pred, label='Regression line', color='red')
plt.scatter(X[outlier_index], y[outlier_index], label='Outlier', color='green')
plt.plot([X[outlier_index], X[outlier_index]], [y[outlier_index], y_pred[outlier_index]], 
         linestyle='--', color='green', label='Residual (Outlier)')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Outlier Detection using Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

#### Interpolation and Extrapolation

```python
import numpy as np
import matplotlib.pyplot as plt

# Known data points
x_known = np.array([1, 2, 3, 4, 5])
y_known = np.array([2, 3, 7, 10, 12])

# Create a line of best fit using numpy's polyfit
fit_coeffs = np.polyfit(x_known, y_known, deg=1)
fit_line = np.poly1d(fit_coeffs)

# Interpolation: Predicting missing data points
x_interpolate = np.array([2.5, 3.5])
y_interpolate = fit_line(x_interpolate)

# Extrapolation: Predicting beyond existing data
x_extrapolate = np.array([0, 6])
y_extrapolate = fit_line(x_extrapolate)

# Create a plot
plt.figure(figsize=(10, 6))

# Plot known data points
plt.scatter(x_known, y_known, color='blue', label='Known Data')

# Plot line of best fit
plt.plot(x_known, fit_line(x_known), color='orange', label='Line of Best Fit')

# Plot interpolated points
plt.scatter(x_interpolate, y_interpolate, color='green', label='Interpolated Points')

# Plot extrapolated points
plt.scatter(x_extrapolate, y_extrapolate, color='red', label='Extrapolated Points')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Interpolation and Extrapolation with Lines')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

#### Image processing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = r'D:\AI-DATASETS\keras\datasets\images\dog.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image loaded successfully
if image is None:
    print("Error: Image not loaded.")
else:
    # Apply Canny edge detection
    edges = cv2.Canny(image, threshold1=100, threshold2=200)

    # Plot the original image and the edges
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(edges, cmap='gray')
    plt.title('Edges Detected')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
