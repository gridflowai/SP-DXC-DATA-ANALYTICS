### Bootstrap sampling
--------------------

Bootstrapping is about taking random samples from your dataset, with replacement. 

For instance, if you have data [2, 4, 6], 
- one bootstrap sample might be [2, 2, 6] and 
- another might be [4, 6, 2].

- Generate a bootstrap sample from the given data.
    
    - `param data`: List of data points from which the sample will be drawn.
    - `param n`: Size of the bootstrap sample. Default is len(data).
    

```python
import random
```

```python
# Create a function to generate a single bootstrap sample
# A bootstrap sample is created by randomly selecting data points 
# from your original dataset with replacement.
def bootstrap_sample(data):
    sample = [random.choice(data) for _ in data]
    return sample
```

```python
# Create a function to generate multiple bootstrap samples:
# create a function called generate_bootstrap_samples that generates 
# multiple bootstrap samples. 
# You can specify how many samples you want to create.
def generate_bootstrap_samples(data, num_samples):
    bootstrap_samples = [bootstrap_sample(data) for _ in range(num_samples)]
    return bootstrap_samples
```

```python
# Your original dataset
data = [20, 30, 40, 50, 60, 70, 80, 90, 100]

# Number of bootstrap samples you want to generate
num_samples = 1000

# Generate the bootstrap samples
bootstrap_samples = generate_bootstrap_samples(data, num_samples)
```
