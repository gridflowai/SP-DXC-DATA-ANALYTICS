------------------------
#### Sampling and inference
-------------------------

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulating a population of exam scores (normally distributed with a mean of 80 and 
# a standard deviation of 10)
population_mean = 80
population_std  = 10
population_size = 10000
```

```python
population_scores = np.random.normal(loc=population_mean, 
                                     scale=population_std, 
                                     size=population_size)
```

```python
# Selecting a random sample from the population
sample_size = 100
sample = np.random.choice(population_scores, sample_size)
```

```python
# Calculating sample statistics
sample_mean = np.mean(sample)
sample_std  = np.std(sample, ddof=1)  # Using Bessel's correction for sample standard deviation
```

```python
# Displaying results
print(f"Population Mean: {population_mean:.2f}")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")

# Creating a histogram of population and sample scores
plt.hist(population_scores, bins=30,   alpha=0.2, label="Population")
plt.hist(sample,            bins=15,   alpha=0.5, label="Sample")

plt.xlabel("Exam Scores")
plt.ylabel("Frequency")
plt.title("Population vs. Sample Exam Scores")
plt.legend()
plt.show()
```

```python
# Creating a histogram for the population scores
plt.subplot(1,2,1)
plt.hist(population_scores, bins=30, alpha=0.5, color='blue', label="Population")
plt.xlabel("Exam Scores")
plt.ylabel("Frequency")
plt.title("Population Exam Scores")
plt.legend()
plt.show()

# Creating a histogram for the sample scores
plt.subplot(1,2,2)
plt.hist(sample, bins=15, alpha=0.5, color='green', label="Sample")
plt.xlabel("Exam Scores")
plt.ylabel("Frequency")
plt.title("Sample Exam Scores")
plt.legend()
plt.show()
```
