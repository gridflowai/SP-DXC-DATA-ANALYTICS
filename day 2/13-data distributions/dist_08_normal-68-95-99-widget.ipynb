```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parameters of the normal distribution
mean    = 0
std_dev = 1

# Generate random data points from the normal distribution
data = np.random.normal(mean, std_dev, 1000)

# Create a histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue')

# Plot the probability density function
x_range = np.linspace(-5, 5, 100)
plt.plot(x_range, norm.pdf(x_range, mean, std_dev), color='red')

plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

#### Example - 02 
- visualize pdfs for various data points

- generate 200 sample data points with a mean height of adults = 170.18 cms and std dev of 2.54 cms.
- data is normally distributed
- generate a function using Ipywidgets using height as float slider with range = mean +- 7*std deviations
- for every changed value of height in the slider, calculate the pdf of the data point and refresh the plot dynamically

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import ipywidgets as widgets
from ipywidgets import interactive
```

```python
# Generate 200 sample data points with a normal distribution
mean_height    = 170.18
std_dev_height = 2.54
num_samples    = 200
```

```python
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(mean_height, std_dev_height, num_samples)
```

```python
# Function to plot the histogram and PDF dynamically
def plot_distribution(height):
    plt.figure(figsize=(14, 6))

    # Plot the histogram of the sample data
    plt.hist(sample_data, bins=20, density=True, alpha=0.7, color='blue', label='Sample Data')

    # Plot the PDF of the normal distribution
    x = np.linspace(mean_height - 7 * std_dev_height, mean_height + 7 * std_dev_height, 1000)
    pdf_values = norm.pdf(x, mean_height, std_dev_height)
    plt.plot(x, pdf_values, color='red', label='Normal Distribution PDF')

    # Highlight the selected height point
    plt.axvline(x=height, color='green', linestyle='--', linewidth=2, label='Selected Height')

    # Set plot details
    plt.title('Normal Distribution and PDF')
    plt.xlabel('Height (cms)')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Create an interactive widget with a float slider for height
height_slider = widgets.FloatSlider(value=mean_height, 
                                    min=mean_height - 7 * std_dev_height, 
                                    max=mean_height + 7 * std_dev_height, 
                                    step=0.1, 
                                    description='Height')
```

```python
# Use the interactive function to dynamically update the plot
interactive_plot = interactive(plot_distribution, height=height_slider)
output = interactive_plot.children[-1]
output.layout.height = '450px'
interactive_plot
```

```python
# Function to plot the histogram and PDF dynamically
def plot_distribution(height):
    plt.figure(figsize=(14, 4))

    # Plot the histogram of the sample data
    plt.hist(sample_data, bins=20, density=True, alpha=0.7, color='blue', label='Sample Data')

    # Plot the PDF of the normal distribution
    x = np.linspace(mean_height - 7 * std_dev_height, mean_height + 7 * std_dev_height, 1000)
    pdf_values = norm.pdf(x, mean_height, std_dev_height)
    plt.plot(x, pdf_values, color='red', label='Normal Distribution PDF')

    # Highlight the selected height point
    plt.axvline(x=height, color='green', linestyle='--', linewidth=2, label='Selected Height')

    # Draw vertical lines for +- 1, 2, and 3 sigma in BLACK
    for i in range(-3, 4):
        sigma_line = mean_height + i * std_dev_height
        plt.axvline(x=sigma_line, color='gray', linestyle='--', alpha=0.5, linewidth=2)

    # Set plot details
    plt.title('Normal Distribution and PDF')
    plt.xlabel('Height (cms)')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

```

```python
# Create an interactive widget with a float slider for height
height_slider = widgets.FloatSlider(value=mean_height, 
                                    min=mean_height - 7 * std_dev_height, 
                                    max=mean_height + 7 * std_dev_height, 
                                    step=0.1, 
                                    description='Height')
```

```python
# Use the interactive function to dynamically update the plot
interactive_plot = interactive(plot_distribution, height=height_slider)
output = interactive_plot.children[-1]
output.layout.height = '450px'
interactive_plot
```

#### calculate pdf using numpy, scipy

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generate data points
x = np.linspace(-3, 3, 1000)
```

```python
# Calculate the PDF for a normal distribution (mean=0, std=1)
pdf_values = norm.pdf(x, loc=0, scale=1)
```

```python
# Plot the PDF
plt.plot(x, pdf_values, label='Normal Distribution PDF')
plt.title('Probability Density Function (PDF) of a Normal Distribution')
plt.xlabel('X')
plt.ylabel('PDF')
plt.legend();
```

- x represents a range of values along the x-axis.
- norm.pdf(x, loc=0, scale=1) calculates the PDF values for a normal distribution with a mean (loc) of 0 and a standard deviation (scale) of 1.
- The plot displays the PDF curve for the normal distribution.
