#### Probability mass function
--------------------------

```python
import numpy as np
```

```python
def pmf(values, probabilities):
    pmf_dict = {}
    for val, prob in zip(values, probabilities):
        pmf_dict[val] = prob
    return pmf_dict

def main():
    # Discrete random variable values and their associated probabilities
    values        = [1, 2, 3, 4, 5, 6]
    probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

    # Calculate the PMF using the custom function
    pmf_result = pmf(values, probabilities)

    # Print the PMF
    print("Probability Mass Function (PMF) for the die roll:")
    for val, prob in pmf_result.items():
        print(f"Value: {val}, Probability: {prob}")


main()
```

**Example 02**

When tossing a fair coin twice, the possible outcomes are {HH, HT, TH, TT}, where H represents a head, and T represents a tail.

Let X be the random variable representing the number of heads observed. The values that X can take are {0, 1, 2}, as there can be 0, 1, or 2 heads in two coin tosses.

Range of X $R_X$: The range of X is the set of all possible values that the random variable X can take. In this case, the range of X is {0, 1, 2}.

`Probability Mass Function (PMF) PX`:
The PMF gives the probabilities associated with each value of the random variable X. Since we have a fair coin (equal probability of getting heads or tails), the probability of getting a head or a tail in a single toss is 0.5.

Let's calculate the PMF for X:

- P(X = 0): The probability of getting 0 heads in two tosses (TT) is 0.5 * 0.5 = 0.25.
- P(X = 1): The probability of getting 1 head in two tosses (HT, TH) is 0.5 * 0.5 + 0.5 * 0.5 = 0.5.
- P(X = 2): The probability of getting 2 heads in two tosses (HH) is 0.5 * 0.5 = 0.25.

So, the PMF for X is:

- PX(0) = 0.25
- PX(1) = 0.5
- PX(2) = 0.25

This means there is a 25% chance of getting 0 heads, a 50% chance of getting 1 head, and a 25% chance of getting 2 heads in two coin tosses.

```python
import matplotlib.pyplot as plt
```

```python
def pmf(number_of_tosses):
    outcomes = ['HH', 'HT', 'TH', 'TT']
    num_heads = [outcome.count('H') for outcome in outcomes]
    total_outcomes = len(outcomes)
    probabilities = [num_heads.count(x) / total_outcomes for x in range(number_of_tosses + 1)]

    pmf_dict = dict(zip(range(number_of_tosses + 1), probabilities))
    return pmf_dict
```

```python
def plot_pmf(pmf_dict):
    plt.bar(pmf_dict.keys(), pmf_dict.values(), align='center', alpha=0.7)
    plt.xticks(list(pmf_dict.keys()))
    plt.xlabel('Number of Heads (X)')
    plt.ylabel('Probability (P(X))')
    plt.title('Probability Mass Function (PMF) of Coin Tosses')
    plt.show()
```

```python
def main():
    number_of_tosses = 2
    pmf_result = pmf(number_of_tosses)
    
    print("Probability Mass Function (PMF) for the number of heads in two coin tosses:")
    for num_heads, prob in pmf_result.items():
        print(f"Number of Heads (X = {num_heads}): Probability (P(X = {num_heads})) = {prob:.2f}")

    plot_pmf(pmf_result)

main()
```

#### Example – uniform pmf - Randomized Sampling

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# List of movies
movies = ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E']

# Number of users to recommend movies to
num_users = 10000

# Initialize a dictionary to store the selection counts
selection_counts = {movie: 0 for movie in movies}

# Randomly recommend movies to users using a uniform PMF
for user in range(num_users):
    recommended_movie = np.random.choice(movies)
    selection_counts[recommended_movie] += 1

# Plot the selection counts
plt.bar(selection_counts.keys(), selection_counts.values())
plt.xlabel('Movies')
plt.ylabel('Selection Counts')
plt.title('Random Movie Selection Counts for 10,000 Users')
plt.xticks(rotation=45)
plt.show()
```

```python
# Print the selection counts
for movie, count in selection_counts.items():
    print(f"{movie}: {count} selections")
```
