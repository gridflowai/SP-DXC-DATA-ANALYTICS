```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set a common seed for reproducibility
np.random.seed(42)
```

```python
# 1. Normal Distribution
data_normal = np.random.normal(size=1000)
plt.figure(figsize=(5, 5))
sns.histplot(data_normal, kde=True)
plt.title('Normal Distribution - 1000 Samples, Range: {:.2f} to {:.2f}'.format(data_normal.min(), data_normal.max()))
plt.xlabel('Values')
plt.ylabel('Frequency');
```

The normal distribution is symmetrical and commonly seen in nature, representing phenomena like height or test scores.

```python
# 2. Uniform Distribution
data_uniform = np.random.uniform(size=1000)
plt.figure(figsize=(5, 5))
sns.histplot(data_uniform, kde=True)
plt.title('Uniform Distribution - 1000 Samples, Range: {:.2f} to {:.2f}'.format(data_uniform.min(), data_uniform.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

The uniform distribution has equal probability for all values within a specified range, often seen in random processes.

```python
# 3. Exponential Distribution
data_exponential = np.random.exponential(size=1000)
plt.figure(figsize=(5, 5))
sns.histplot(data_exponential, kde=True)
plt.title('Exponential Distribution - 1000 Samples, Range: {:.2f} to {:.2f}'.format(data_exponential.min(), data_exponential.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

The exponential distribution models the time between independent events, often used in reliability and queuing theory.

```python
# 4. Binomial Distribution
data_binomial = np.random.binomial(n=10, p=0.5, size=1000)
plt.figure(figsize=(5, 5))
sns.histplot(data_binomial, kde=True)
plt.title('Binomial Distribution - 1000 Samples, Range: {:.2f} to {:.2f}'.format(data_binomial.min(), data_binomial.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials.

```python
# 5. Poisson Distribution
data_poisson = np.random.poisson(lam=3, size=1000)
plt.figure(figsize=(5, 5))
sns.histplot(data_poisson, kde=True)
plt.title('Poisson Distribution - 1000 Samples, Range: {:.2f} to {:.2f}'.format(data_poisson.min(), data_poisson.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

The Poisson distribution models the number of events occurring in a fixed interval of time or space.

```python
# 11. Left-Skewed Distribution (Negatively Skewed)
data_left_skewed = np.random.normal(loc=10, scale=5, size=1000)
data_left_skewed = np.concatenate([data_left_skewed, np.random.normal(loc=20, scale=1, size=200)])

plt.figure(figsize=(5, 5))
sns.histplot(data_left_skewed, kde=True)
plt.title('Left-Skewed Distribution - 1200 Samples, Range: {:.2f} to {:.2f}'.format(data_left_skewed.min(), data_left_skewed.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

Left-skewed distributions, also known as negatively skewed, have a tail extending to the left. Common in income distribution where high earners pull the mean to the right.

```python
data_right_skewed = np.random.exponential(scale=2, size=1000)
data_right_skewed = np.concatenate([data_right_skewed, np.random.exponential(scale=8, size=200)])

plt.figure(figsize=(5, 5))
sns.histplot(data_right_skewed, kde=True)
plt.title('Right-Skewed Distribution - 1200 Samples, Range: {:.2f} to {:.2f}'.format(data_right_skewed.min(), data_right_skewed.max()))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

Right-skewed distributions, also known as positively skewed, have a tail extending to the right. Common in lifetimes of products where early failures occur.
