---------------------
### Asymmetry of data
-----------------------

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # For density plots
```

```python
# Generate skewed data
np.random.seed(0)
skewed_data = np.random.exponential(scale=2, size=1000) + 5
```

```python
# Create a histogram
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.hist(skewed_data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram')

# Create a box plot
plt.subplot(1, 3, 2)
plt.boxplot(skewed_data, vert=False)
plt.title('Box Plot')

# Create a density plot
plt.subplot(1, 3, 3)
sns.kdeplot(skewed_data, fill=True)
plt.title('Density Plot')

plt.tight_layout()
```

```python
# Generate left-skewed data
np.random.seed(0)
left_skewed_data = -np.random.exponential(scale=2, size=1000) + 5

# Create a histogram
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.hist(left_skewed_data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram')

# Create a box plot
plt.subplot(1, 3, 2)
plt.boxplot(left_skewed_data, vert=False)
plt.title('Box Plot')

# Create a density plot
plt.subplot(1, 3, 3)
sns.kdeplot(left_skewed_data, shade=True)
plt.title('Density Plot')

plt.tight_layout()
plt.show()
```

#### income distribution, 

- which is typically right-skewed because a small percentage of the population earns significantly higher incomes compared to the majority.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate synthetic income data with right-skew
np.random.seed(42)
income_data = np.random.exponential(scale=3000, size=1000)
```

```python
# Visualize the income distribution
plt.figure(figsize=(8, 6))
sns.histplot(income_data, kde=True)
plt.title('Income Distribution (Right-skewed)')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

#### Wealth Distribution (Right-skewed): 

- Just like income distribution, wealth distribution is often right-skewed, where a small percentage of the population holds a significant portion of the total wealth.

```python
# Generate synthetic wealth data with right-skew
np.random.seed(42)
wealth_data = np.random.exponential(scale=500000, size=1000)

# Visualize the wealth distribution
plt.figure(figsize=(8, 6))
sns.histplot(wealth_data, kde=True)
plt.title('Wealth Distribution (Right-skewed)')
plt.xlabel('Wealth')
plt.ylabel('Frequency')
```

##### Company Profits (Right-skewed): 

- Company profit distributions can also be right-skewed, as a few successful companies might earn significantly higher profits compared to the majority.

```python
# Generate synthetic company profit data with right-skew
np.random.seed(42)
profit_data = np.random.exponential(scale=1000000, size=1000)

# Visualize the company profit distribution
plt.figure(figsize=(8, 6))
sns.histplot(profit_data, kde=True)
plt.title('Company Profit Distribution (Right-skewed)')
plt.xlabel('Profit')
plt.ylabel('Frequency');
```

##### Inflation Rates (Left-skewed): 

- Inflation rates often exhibit left-skewness because extreme deflation is rarer compared to moderate inflation.

```python
# Generate synthetic inflation rate data with left-skew
np.random.seed(42)
inflation_data = -np.random.exponential(scale=1, size=1000)

# Visualize the inflation rate distribution
plt.figure(figsize=(8, 6))
sns.histplot(inflation_data, kde=True)
plt.title('Inflation Rate Distribution (Left-skewed)')
plt.xlabel('Inflation Rate')
plt.ylabel('Frequency')
```
