-----------------------
#### concept of independence in the context of t-tests
---------------------------------

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate dummy data
np.random.seed(0)
```

```python
# Independent Scenario: Heights from two different schools
school_1_heights = np.random.normal(150, 10, 50)  # 150cm mean height, 10cm standard deviation
school_2_heights = np.random.normal(180, 3,  50)  
```

```python
# Dependent Scenario: Heights from two families (with some correlation among family members)
base_heights_1 = np.random.normal(150, 10, 50)
base_heights_2 = np.random.normal(155, 3,  50)

family_1_heights = base_heights_1 + np.random.normal(0, 2, 50)  # siblings deviate a bit from the base height
family_2_heights = base_heights_2 + np.random.normal(5, 2, 50)  # siblings deviate a bit more, taller family
```

```python
plt.figure(figsize=(12,5))
# Independent Samples from Two Schools
plt.subplot(1, 2, 1)
plt.hist(school_1_heights, alpha=0.5, label='School 1', bins=10, color='blue', edgecolor='black')
plt.hist(school_2_heights, alpha=0.5, label='School 2', bins=10, color='red',  edgecolor='black')
plt.title('Independent Samples from Two Schools', fontsize=10)
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.legend()

# Dependent Samples from Two Families
plt.subplot(1, 2, 2)
plt.hist(family_1_heights, alpha=0.5, label='Family 1', bins=10, color='blue', edgecolor='black')
plt.hist(family_2_heights, alpha=0.5, label='Family 2', bins=10, color='red',  edgecolor='black')
plt.title('Dependent Samples from Two Families', fontsize=10)
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout();
```

```python
# Plotting
plt.figure(figsize=(12, 5))

# Independent Samples from Two Schools
plt.subplot(1, 2, 1)
plt.boxplot([school_1_heights, school_2_heights], labels=['School 1', 'School 2'])
plt.title('Independent Samples from Two Schools')
plt.ylabel('Height (cm)')

# Dependent Samples from Two Families
plt.subplot(1, 2, 2)
plt.boxplot([family_1_heights, family_2_heights], labels=['Family 1', 'Family 2'])
plt.title('Dependent Samples from Two Families')
plt.ylabel('Height (cm)')

plt.tight_layout()
```
