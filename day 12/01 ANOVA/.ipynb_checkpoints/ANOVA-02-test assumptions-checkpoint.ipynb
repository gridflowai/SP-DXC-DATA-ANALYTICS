----------------------
#### Assumptions for ANOVA
--------------------------

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Set a seed for reproducibility
np.random.seed(42)
```

```python
# Generate larger samples for three groups
group1 = np.random.normal(loc=30, scale=5, size=25)
group2 = np.random.normal(loc=35, scale=5, size=25)
group3 = np.random.normal(loc=40, scale=5, size=25)
```

```python
# Perform Shapiro-Wilk test for normality for each group
shapiro_stat_group1, shapiro_p_value_group1 = stats.shapiro(group1)
shapiro_stat_group2, shapiro_p_value_group2 = stats.shapiro(group2)
shapiro_stat_group3, shapiro_p_value_group3 = stats.shapiro(group3)
```

```python
# Print the results
print("Shapiro-Wilk Test for Normality (Group 1):")
print(f"Test Statistic: {shapiro_stat_group1}")
print(f"P-value: {shapiro_p_value_group1}")
print("\nShapiro-Wilk Test for Normality (Group 2):")
print(f"Test Statistic: {shapiro_stat_group2}")
print(f"P-value: {shapiro_p_value_group2}")
print("\nShapiro-Wilk Test for Normality (Group 3):")
print(f"Test Statistic: {shapiro_stat_group3}")
print(f"P-value: {shapiro_p_value_group3}")
```

 If the p-value from the test is greater than the chosen significance level (e.g., 0.05), you might consider the group's data as approximately normally distributed.

```python
# Create Q-Q plots for visual inspection of normality for each group
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
stats.probplot(group1, dist="norm", plot=plt)
plt.title("Q-Q Plot (Group 1)")

plt.subplot(1, 3, 2)
stats.probplot(group2, dist="norm", plot=plt)
plt.title("Q-Q Plot (Group 2)")

plt.subplot(1, 3, 3)
stats.probplot(group3, dist="norm", plot=plt)
plt.title("Q-Q Plot (Group 3)")
```
